# Introduccion a la libreria dplyr

Cuando trabajamos con datos debemos tener claro nuestro objetivo.
En este sentido, ayuda mucho  el escribir las tareas en la forma de un algoritmo para luego simplemente ejecutar este algoritmo. La libreria dplyr nos permite realizar esto de manera rapida y simple.

 Esta libreria provee de verbos simples que nos permiten realizar tareas de manipulacion de datos. Cada funcion esta programada de manera eficiente de modo de optimizar el  uso de tiempo y otros recursos.

### A modo de ejemplo, trabajaremos con el conjunto de datos nycflights13

```r
#install.packages("nycflights13")
library(nycflights13)
library(dplyr)
library(ggplot2)
dim(flights)
flights
```

Es importante notar que esta tabla es de tipo tibble, una reinvencion moderna de los data frames lista de verbos

* filter() permite seleccionar casos basados en valores u operadores logicos
* arrange() permite reordenar los casos
* select() y rename() para seleccionar variables basadas en sus nombres
* mutate() y transmute() para agregar nuevas variables que sean funciones de las ya existentes
* summarise() para condensar varios valores en un unico valor
* sample_n() y sample_frac() para tomar muestras aleatorias

# Operaciones basadas en patrones

Operaciones agrupadas:
 Los verbos contenidos en la librería dplyr son muy útiles pero su utilidad se potencia cuando estos son mezclados con la posibilidad de agrupar registros. 

Revisemos un ejemplo, separaremos el conjunto de datos por cada avión y contaremos el número de vuelos, la distancia media recorrida, el promedio de retraso y graficaremos los resultados

```r
by_tailnum <- dplyr::group_by(flights, tailnum)
delay <- dplyr::summarise(by_tailnum,
                   count = n(),
                   dist = mean(distance, na.rm = TRUE),
                   delay = mean(arr_delay, na.rm = TRUE))
delay <- dplyr::filter(delay, count > 20, dist < 2000)
```

```r
ggplot(delay, aes(dist, delay)) +
  geom_point(aes(size = count), alpha = 1/2) +
  geom_smooth() +
  scale_size_area()
```

Es curioso que el retraso promedio no parece tener relacion directa con la distancia media recorrida Cuando usamos summarise() en conjunto con medidas de agregacion, la medida se aplica sobre un vector y nos entrega un valor numerico. Hay varias funciones en R que podemos utilizar, min(), max(), sd(), etc. 

 La libreria dplyr trae incluidas ademas algunas extras

* n(): cuenta el numero de registros en cada grupo
* n_distinct(x): cuenta el numero de valores distintos de x
* first(x), last(x) y nth(x, n) recatan el primer, ultimo y enesimo valor de x

# Operaciones de mutación

### La semántica de las operaciones de mutacion es diferente de la semantica de las de seleccion.

* select() requiere de nombres o posiciones, en cambio mutate recibe vectores columna.

```r
df <- select(flights, year:dep_time)
df
```

### Cuando utilizamos select(), los nombre de columnas vacios hacen referencia a la posición de la columna en el arreglo tibble. Para mutate(), el símbolo representa al valor almacenado en la columna. 

### Consideremos lo que ocurre si pasamos una cadena de caracteres o un numero a mutate()


```r
mutate(df, "year", 2)
```

### notemos que los arreglos recibidos son reciclados de modo de calzar con el número de filas

```r
mutate(df, year + 10)
```

### podemos además crear nombres nuevos en la creación de la columna

```r
var <- seq(1, nrow(df))
mutate(df, new = var)
```
